```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/zakg04/Documents/HBC Lab/BOOST/Main/test-7005/FN/7005_FN_A.csv')

learn = df[df['block']=='learn']
test = df[df['block']=='test']

test
```

```python
# plot percent correct for test on a bar graph 0-1

plt.figure(figsize=(3,5))
sns.barplot(y='correct', data=test)
plt.title('Test Percent Correct')
#add percentage correct above bar
percentage = test['correct']/len(test)
plt.text(0, 0.5, str(1 - percentage.mean()) + '%', ha='center', va='center')
plt.ylabel('Percent Correct')
plt.xlabel('Condition')
plt.ylim(0,1)
cap_text = 'n = ' + str(len(test)) + '\n' + 'Total Correct = ' + str(test['correct'].sum())
plt.text(0, -0.2, cap_text, ha='center', va='center')
```

```python
#dot plot of reaction time for test with box and whiskers

plt.figure(figsize=(5,5))
sns.boxplot(y='response_time', data=test)
sns.swarmplot(y='response_time', data=test, color='black')
plt.title('Test Reaction Time')
plt.ylabel('Reaction Time')
plt.xlabel('Condition')
cap_text = 'n = ' + str(len(test)) + '\n' + 'Mean RT = ' + str(test['response_time'].mean()) 
plt.text(0, -0.01, cap_text, ha='center', va='center')
plt.show()
```
